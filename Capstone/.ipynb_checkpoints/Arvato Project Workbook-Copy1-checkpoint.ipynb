## Functions used

### clean_data

```python
def clean_data(df, columns_to_use, remove_nulls = True):
    '''
    Info:
        This function cleans the dataframe, reducing the amount of features used and removing/replacing the samples with 
        null values.
    ----------
    Input:
        df: DataFrame to be cleaned (type: pandas DataFrame)
        columns_to_use: List of columns to be used (type: list)
        remove_nulls: Boolean variable to decide if the samples with null values shall be removed (True) or if the null
        values shall be replaced with the mode of the column (False)
    ----------
    Output:
        df_clean: DataFrame after the cleaning steps (type: pandas DataFrame)
    '''
    
    df_dirty = df.copy()
    
    # Filtering the columns to be used
    df_dirty = df_dirty[columns_to_use]
    
    if remove_nulls:
        # Removing the samples with null values
        df_clean = df_dirty.dropna()
    else:
        
        for col in columns_to_use:
            # Replace the null values with the most common value of the column
            df_dirty[col] = df_dirty[col].fillna(df_dirty[col].mode()[0])
        
    
    return df_dirty
```

### numeric_df

```python
def numeric_df(df):
    '''
    Info:
        This function removes non numeric columns from the dataframe.
    ----------
    Input:
        df: Dataframe to be filtered (type: pandas DataFrame)
    ----------
    Output:
        df_num: Dataframe containing only numeric columns (type: pandas DataFrame)
        person_info: Column with the information if the person is from the customers or azdias dataframes
    '''
    
    df_num = df.copy()
    
    # Saving the column with the information if the person is from the customers or azdias dataframes
    person_info = df_num['Dataframe'].values
    
    # Obtaining only numeric data
    numeric_cols = list(df_num.select_dtypes(include = 'number').columns)
    print('Number of numeric columns: ', len(numeric_cols))

    df_num = df_num[numeric_cols]
    
    return df_num, person_info

```

### calculate_WSS

```python
def calculate_WSS(points, kmax):
    '''
    Info:
        This function returns the WSS score for k values from 1 to kmax.
    ----------
    Inputs:
        points: Points of the data to be clustered (type: numpy array)
        kmax: Maximum value of k to be analysed (type: integer)
    ----------
    Outputs:
        see: WSS score for k values from 1 to kmax (type: list)
    '''
    
    sse = []
    for k in range(1, kmax+1):
        kmeans = KMeans(n_clusters = k, random_state = 0).fit(points)
        centroids = kmeans.cluster_centers_
        pred_clusters = kmeans.predict(points)
        curr_sse = 0
    
        # calculate square of Euclidean distance of each point from its cluster center and add to current WSS
        for i in range(len(points)):
            curr_center = centroids[pred_clusters[i]]
            curr_sse += (points[i, 0] - curr_center[0]) ** 2 + (points[i, 1] - curr_center[1]) ** 2
      
        sse.append(curr_sse)
        
    plt.plot(sse)
    plt.title('WSS x K')
    plt.xlabel('K')
    plt.ylabel('WSS');
    
    return sse
```

### models_prediction

```python
def models_prediction(df):
    '''
    Info:
        This functions trains and tests several models and returns the results of the predictions of each one. The models are:
        - Sckit-learn Multi Layer Perceptron
        - Random Forest Classifier
        - Logistic Regression
        - Support Vector Machine
    ----------
    Input:
        df: DataFrame used to train and test the models (type: pandas DataFrame)
    ----------
    Output:
        y_test: Sample of the labels used in the testing of the models (type: pandas Series)
        y_predicted: Dictionary containing the predictions made by each model (type: dict)
    '''
    
    # Applying SMOTE
    sm = SMOTE(random_state = 0)
    X_res, y_res = sm.fit_resample(df.drop(columns = 'RESPONSE'), df['RESPONSE'])
    
    # Splitting the data into training (80%) and test (20%) sets
    X_train, X_test, y_train, y_test = train_test_split(X_res, 
                                                        y_res, 
                                                        test_size = 0.2,
                                                        random_state = 0,
                                                        stratify = y_res
                                                        )

    # Initializing the models with default parameters
    mlp = MLPClassifier(random_state = 0, 
        activation='relu')
    
    rf = RandomForestClassifier(random_state = 0)
    
    lr = LogisticRegression(random_state=0)
    
    svm = SVC(random_state = 0)
    
    # Training and testing the models
    print("Training the MLP")
    mlp.fit(X_train, y_train)
    y_mlp = mlp.predict(X_test)
    
    print("\nTraining the Random Forest Classifier")
    rf.fit(X_train, y_train)
    y_rf = rf.predict(X_test)
    
    print("\nTraining the Logistic Regressor")
    lr.fit(X_train, y_train)
    y_lr = lr.predict(X_test)
    
    print("\nTraining the Support Vector Machine")
    svm.fit(X_train, y_train)
    y_svm = svm.predict(X_test)
    
    # Saving the predictions in a dictionary
    y_predicted = {'mlp': y_mlp, 'rf': y_rf, 'lr': y_lr, 'svm': y_svm}    
    
    return y_test, y_predicted
```

### plot_confusion

```python
def plot_confusion(cm, acc):
    '''
    Info:
        This function plots the confusion matrix
    ----------
    Input:
        cm: Confusion matrix (type:)
        acc: Accuracy of the model (type: float)
    ----------
    Output:
        None
    '''
    # Creating the figure
    plt.figure(figsize=(5,5))
    
    # Ploting the Confusion Matrix
    sns.heatmap(cm, annot=True, fmt=".0f", linewidths=.5, square = True, cmap = 'Blues_r');
    
    # Setting the axis labels
    plt.ylabel('Actual label');
    plt.xlabel('Predicted label');
    
    # Setting the title for the figure
    all_sample_title = 'Test Accuracy Score: {0:.3f}'.format(acc)
    plt.title(all_sample_title, size = 15);
```

### calculate_metrics

```python
def calculate_metrics(y_true, y_pred):
    '''
    Info:
        This function calculates the accuracy, precision, recall, ROC AUC and Confusion Matrix.
    ----------
    Input:
        y_true: True value of the labels (type: numpy array)
        y_pred: Predicted values of the labels (type: numpy array)
    ----------
    Output:
        acc: Accuracy calculated (type: float)
        prec: Accuracy calculated (type: float)
        rec: Accuracy calculated (type: float)
        auc: Accuracy calculated (type: float)
        cm: Confusion Matrix (type: numpy array)
    '''
    
    # Calculating the Accuracy
    acc = accuracy_score(y_true, y_pred)
    
    # Calculating the Precision
    prec = precision_score(y_true, y_pred)
    
    # Calculating the Recall
    rec = recall_score(y_true, y_pred)
    
    # Calculating the ROC AUC
    auc = roc_auc_score(y_true, y_pred)
    
    # Calculating the Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    
    return acc, prec, rec, auc, cm
```

Each of the rows in the "MAILOUT" data files represents an individual that was targeted for a mailout campaign. Ideally, we should be able to use the demographic information from each individual to decide whether or not it will be worth it to include that person in the campaign.

The "MAILOUT" data has been split into two approximately equal parts, each with almost 43 000 data rows. In this part, you can verify your model with the "TRAIN" partition, which includes a column, "RESPONSE", that states whether or not a person became a customer of the company following the campaign. In the next part, you'll need to create predictions on the "TEST" partition, where the "RESPONSE" column has been withheld.

## Hyperparameter tunning

To find the best values for the hyperparameters of the Random Forest Classifier model, the Optuna library will be used. 

Optuna is an open-source python library for hyperparameter optimization with a very simple and intuitive implementation. 

```python
class Objective_LSTM:

    def __init__(self):
        self.best_model = None
        self._model = None
        
    def __call__(self, trial):
        


        model.fit(
            x_train,
            y_train,
            validation_data=(x_test, y_test),
            callbacks=[pruning_callback, early_stop_callback],
            shuffle=False,
            batch_size=trial.suggest_int("batch_size", 3, 8),
            epochs=100,
            verbose=2)
        
        self._model = model

        # Evaluate the model accuracy on the validation set.
        score = model.evaluate(x_test, y_test, verbose=0)
        return score
    
    def callback(self, study, trial):
        
        if study.best_trial == trial:
            self.best_model = self._model
```

```python
class Opt_RF:
    
    def __init__(self):
        self.best_model = None
        self._model = None
        
    def __call__(self, trial):
        # Definition of the search space
        params = {
            'n_estimators': trial.suggest_int('n_estimators', 10, 1000),
            'criterion': trial.suggest_categorical('criterion', ['gini', 'entropy']),
            #'max_depth': trial.suggest_int('max_depth', 2, 10),
            'min_samples_split': trial.suggest_int('min_samples_split', 2, 100),
            'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 100)
        }

        # Initializing the model
        model = RandomForestClassifier(**params, random_state=0)

        # Training the model
        model.fit(X_train, y_train)
        
        self._model = model

        # Testing
        y_pred = model.predict(X_test)

        # Calculating the ROC AUC as evaluating metric
        auc = roc_auc_score(y_test, y_pred)

        return auc
    
    def callback(self, study, trial):
        
        if study.best_trial == trial:
            self.best_model = self._model
```

```python
# Splitting the data into training (80%) and test (20%) sets
X_train, X_test, y_train, y_test = train_test_split(X2_res, 
                                                    y2_res, 
                                                    test_size = 0.2,
                                                    random_state = 0,
                                                    stratify = y2_res
                                                    )
objective = Opt_RF()

# Creating the study
study = optuna.create_study(study_name='Random_Forest_Opt', 
                            direction="maximize", 
                            storage='sqlite:///opt_saved_trials.db',
                            load_if_exists = True)

# Begin the optimization process
study.optimize(objective, n_trials = 100, callbacks = [objective.callback])
```

```python
# DOWNSAMP
# Separate input features and target
y = df2_mailout['RESPONSE']
X = df2_mailout.drop('RESPONSE', axis=1)

# setting up testing and training sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# concatenate our training data back together
X = pd.concat([X_train, y_train], axis=1)

print('Distribuio original:')
print(X.RESPONSE.value_counts())

# separate minority and majority classes
not_mail = X[X.RESPONSE==0]
mail = X[X.RESPONSE==1]

# downsample majority
not_mail_downsampled = resample(not_mail,
                          replace=False, # sample without replacement
                          n_samples=round(len(not_mail)/2), 
                          random_state = 0) # reproducible results

# combine majority and upsampled minority
downsampled = pd.concat([not_mail_downsampled, mail])

print('Distribuio downsampled:')
print(downsampled.RESPONSE.value_counts())

# UPSAMP
# separate minority and majority classes
not_mail2 = downsampled[downsampled.RESPONSE==0]
mail2 = downsampled[downsampled.RESPONSE==1]

# upsample minority
not_mail_upsample = resample(mail2,
                          replace=True, # sample without replacement
                          n_samples=len(not_mail2), 
                          random_state = 0) # reproducible results

# combine majority and upsampled minority
upsampled = pd.concat([not_mail_upsample, not_mail2])

print('Distribuio upsampled:')
print(upsampled.RESPONSE.value_counts())
```

```python
y_tr = upsampled['RESPONSE']
X_tr = upsampled.drop('RESPONSE', axis=1)

model = RandomForestClassifier(random_state = 0, class_weight = 'balanced').fit(X_tr, y_tr)

y_pred = model.predict(X_test)

# Checking accuracy
print(accuracy_score(y_test, y_pred))
    
print(recall_score(y_test, y_pred))

print(roc_auc_score(y_test, y_pred))
```
